```python
import bs4 as bs
import urllib.request
import re
```

```python
source = urllib.request.urlopen('https://en.wikipedia.org/wiki/Global_warming').read()
```

```python
soup = bs.BeautifulSoup(source,"html.parser")
print(soup)
text = ""
```

```python
for paragraph in soup.find_all('p'):
    text += paragraph.text
text
text = "A defiant President Trump pushed back Thursday against the outrage over his comments that he would be open to accepting opposition research from individuals from foreign countries, arguing he should have no obligation to call the FBI in certain cases while trying to turn the tables on Democratic lawmakers over their own foreign contacts.I meet and talk to ‘foreign governments’ every day, the president tweeted Thursday.Citing recent conversations with leaders in the United Kingdom and Ireland, France and Poland, Trump said, We talked about ‘Everything!’ Should I immediately call the FBI about these calls and meetings? How ridiculous! I would never be trusted again. He initially included a widely mocked spelling error referencing the Prince of Whales but later corrected it."
```

```python
text = re.sub(r"[\[0-9]*\]",' ',text)
text = re.sub(r"\s+",' ',text)
text
# text  = nltk.sent_tokenize(text)
```

```python
clean_text = text.lower()
clean_text = re.sub(r'\W',' ',clean_text)
clean_text = re.sub('\d',' ',clean_text)
clean_text = re.sub('\s+',' ',clean_text)
clean_text
```

```python
import nltk 
sentences = nltk.sent_tokenize(text)
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
stopwords
```

```python
word2count = {}
for word in nltk.word_tokenize(clean_text):
    if word not in stopwords:
        if word not in word2count:
            word2count[word] = 1
        else:
            word2count[word] += 1
word2count
```

```python
for key in word2count.keys():
    word2count[key] = word2count[key]/max(word2count.values())
    #print(key)
#word2count['air']
word2count.values()
```

```python
sent2score = {}
for sentence in sentences:
    for word in nltk.word_tokenize(sentence.lower()):
        if word in word2count.keys():
            if len(sentence.split()) < 30:
                if sentence not in sent2score.keys():
                    sent2score[sentence] = word2count[word]
                else:
                    sent2score[sentence] += word2count[word]
print(sent2score)
max(sent2score.values())
```

```python
import heapq
```

```python
best_sentences = heapq.nlargest(10,sent2score,key=sent2score.get)
```

```python
best_sentences
```

```python
for sentence in best_sentences:
    print(sentence)
```
